# BIVARIATE / MULTIVARIATE ANALYSIS 

```python
# SCATTER PLOT - ( NUMERICAL - NUMERICAL )
sns.scatterplot(x = df_clean["EngineSize"],
                y = df_clean["Horsepower"])
```

```python
sns.scatterplot(x = df_clean['Length'],
                y = df_clean['Wheelbase'])
```

```python
df_clean.columns
```

```python
# BARPLOT ( NUMERICAL - CATEGORICAL )
sns.barplot(x = df_clean['Type'],
            y = df_clean['Cylinders'])
```

```python
# BOXPLOT(NUMERICAL-CATEGORICAL) # we can aslo check outliers using this 
sns.boxplot(x = df_clean['Type'],
            y = df_clean['Cylinders'])
```

```python
#HEATMAP(CATEGORICAL - CATEGORICAL) 
# FOR USING HEATMAP WE HAVE TO CREATE A HEAT MAP FIRST THEN WE CAN PROCEED 
cross = pd.crosstab( df_clean['Type'],
                    df_clean['DriveTrain'])
sns.heatmap(cross,annot=True, cmap='Blues')
```

```python
#CLUSTER MAP ( ADVANCED VERSION OF HEATMAP ) 
corr = df_clean.corr(numeric_only=True)

sns.clustermap(corr,
               annot=True,
               cmap='coolwarm')

```

```python
table = pd.crosstab(df_clean['Type'],
                    df_clean['DriveTrain'])

sns.clustermap(table,
               annot=True,
               cmap='Blues')

```
